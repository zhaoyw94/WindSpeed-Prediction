```python
import sqlite3
import pandas as pd 
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
data = pd.read_csv('daily_weather.csv')
```

```python
data.head(10)
```

```python
data.shape
```

```python
data.columns
```

```python
features = ['air_pressure_9am', 'air_temp_9am', 'avg_wind_direction_9am',
       'avg_wind_speed_9am', 'max_wind_direction_9am', 'rain_duration_9am',
       'rain_accumulation_9am']

target = ['max_wind_speed_9am']
```

```python
data = data.dropna()
data.shape
```

```python
X = data[features]
y = data[target]
```

** Let us look at a typical row from our features: **

```python
X.iloc[334]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=324)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_prediction = regressor.predict(X_test)
y_prediction
```

```python
y_test.describe()
```

```python
RMSE = sqrt(mean_squared_error(y_true = y_test, y_pred = y_prediction))
print(RMSE)
```

** Using decision tree model to see RMSE

```python
regressor = DecisionTreeRegressor(max_depth=20)
regressor.fit(X_train, y_train)
```

```python
y_prediction = regressor.predict(X_test)
y_prediction
```

```python
y_test.describe()
```

```python
RMSE = sqrt(mean_squared_error(y_true = y_test, y_pred = y_prediction))
print(RMSE)
```
